```python
#Reading data using pandas 
import pandas as pd
df = pd.read_csv("orders.csv")
df.head(10)
```

```python
#Treating unavailable values as Null(NaN) 
df = pd.read_csv("orders.csv", na_values= ['Not Available', 'unknown']) #Shift + Tab 
df.head(10)
```

```python
#fixing column names of this dataframe 
df.columns = df.columns.str.lower()
df.head()
```

```python
df.columns = df.columns.str.replace(' ','_')
df.head()
```

```python
#Deriving new columns 
df['discount'] = df['list_price']*df['discount_percent']*0.01 #To get the exact discount price
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price'] 
df.head()
```

```python
#Convert OrderDate type from Object to Date 
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.head()
```

```python
#Dropping unnecessary columns
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace = True)
df.head()
```

```python
#Load this data into SQL Server 
import sqlalchemy as sal #To connect to sql server 
engine = sal.create_engine('mssql://LenovoIdeaPadSlim3-2024/master?driver=ODBC+DRIVER+18+FOR+SQL+SERVER&TrustServerCertificate=yes') #Creating a connection [sql, server_name, db_name?odbc_driver] 
conn = engine.connect()  #Created the connection #Server:: LENOVOIDEAPADSL
```

```python
#Load data into SQL Server
df.to_sql('df_orders', con=conn, index=False, if_exists = "append") #Columns taking more memory as Varchar(max) & bignit if "replaced"
```
